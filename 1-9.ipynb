```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
n = 1000 #number of random data points 

x = np.random.uniform(0,1,n) #generate random data

#creating the histograms
width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=100, alpha=0.5, edgecolor="black")

#labels
plt.ylabel("N ber bin")
plt.xlabel("x")
plt.savefig("1-9.pdf", bbox_inches="tight",facecolor="white")
```
